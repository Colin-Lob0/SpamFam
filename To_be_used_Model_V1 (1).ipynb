```python
import pandas as pd
msg = pd.read_csv('/content/Final_Dataset.csv')
```

```python
msg
```

```python
import re
import nltk
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
from nltk.corpus import stopwords
```

```python
nltk.download('wordnet')
nltk.download('stopwords')

corpus = []
for i in range(0, len(msg)):
    # Convert to string in case of non-string values
    review = str(msg['Body'][i])

    review = re.sub('[^a-zA-Z]', ' ', review)
    review = review.lower()
    review = review.split()

    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## Tokenization using TF-IDF

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
msg.Label = le.fit_transform(msg.Label)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer(max_features=2500, ngram_range = (1,2))
X = tv.fit_transform(corpus).toarray()
```

```python
y = msg['Label']
y
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = classifier.predict(X_test)
print('Accuracy of Testing :',accuracy_score(y_pred,y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```
